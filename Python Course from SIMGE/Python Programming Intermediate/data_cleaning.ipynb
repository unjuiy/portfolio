```python
import pandas as pd

df = pd.read_csv("data.csv")
df
```

```python
# Remove all invalid entries
#  blank entry

df = df.dropna()
df

## N657, Malay, blank entry
## p2586, Mandarin, blank entry
## p1589, Mandarin, blank entry
```

```python
#  invalid rating entry

#print(df.dtypes)
#df = df.astype({"Q4": 'float'}) # ValueError: could not convert string to float: 'a'

q4_remove_letter = pd.Series(df["Q4"] != 'a')
df = df[q4_remove_letter]
df = df.astype({"Q4": 'float'})
df

## P3462, Tamil, invalid rating entry
```

```python
#  invalid rating range

q1_mask = pd.Series(df["Q1"] >= -2) & pd.Series(df["Q1"] <= 2)
q2_mask = pd.Series(df["Q2"] >= -2) & pd.Series(df["Q2"] <= 2)
q3_mask = pd.Series(df["Q3"] >= -2) & pd.Series(df["Q3"] <= 2)
q4_mask = pd.Series(df["Q4"] >= -2) & pd.Series(df["Q4"] <= 2)
q5_mask = pd.Series(df["Q5"] >= -2) & pd.Series(df["Q5"] <= 2)
q6_mask = pd.Series(df["Q6"] >= -2) & pd.Series(df["Q6"] <= 2)
q7_mask = pd.Series(df["Q7"] >= -2) & pd.Series(df["Q7"] <= 2)
q8_mask = pd.Series(df["Q8"] >= -2) & pd.Series(df["Q8"] <= 2)
mask = q1_mask & q2_mask & q3_mask & q4_mask & q5_mask & q6_mask & q7_mask & q8_mask
df = df[mask]
df

## p2614, Mandarin, invalid rating range
```

```python
# Remove entry with Subject ID “test”

df = df[df["Subject ID"] != "test"]
df
```

```python
# Remove all the other questions response entry

df = df.loc[:,:"Q8"]
df
```

```python
# Store the cleaned data in a new csv file 

df.to_csv("cleaned_data.csv",index=False)
```
