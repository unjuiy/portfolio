```python
import pandas as pd

df = pd.read_csv("cleaned_data.csv")
df
```

```python
# For each mother tongue, compute the 8 questions overall

questions = df[["Q1","Q2","Q3","Q4","Q5","Q6","Q7","Q8"]]

Mandarin = pd.Series(df["Language"] == 'Mandarin')
df_mandarin = questions[Mandarin]

Malay = pd.Series(df["Language"] == 'Malay')
df_malay = questions[Malay]

Tamil = pd.Series(df["Language"] == 'Tamil')
df_tamil = questions[Tamil]

mandarin_mean_for_each_question = df_mandarin.mean()
malay_mean_for_each_question = df_malay.mean()
tamil_mean_for_each_question = df_tamil.mean()
mandarin_std_for_each_question = df_mandarin.std()
malay_std_for_each_question = df_malay.std()
tamil_std_for_each_question = df_tamil.std()

#  Mean
print("Mandarin Mean: ")
print(mandarin_mean_for_each_question.mean() )
print("\nMalay Mean: ")
print(malay_mean_for_each_question.mean() )
print("\nTamil Mean: ")
print(tamil_mean_for_each_question.mean() )

#  Sample Standard Deviation
print("\nMandarin Sample Standard Deviation: ")
print(mandarin_std_for_each_question.std())
print("\nMalay Sample Standard Deviation: ")
print(malay_std_for_each_question.std())
print("\nTamil Sample Standard Deviation: ")
print(tamil_std_for_each_question.std())
```
