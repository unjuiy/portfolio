```python
import pandas as pd

df = pd.read_csv("cleaned_data.csv")
df
```

```python
# For each mother tongue and question, compute

questions = df[["Q1","Q2","Q3","Q4","Q5","Q6","Q7","Q8"]]

Mandarin = pd.Series(df["Language"] == 'Mandarin')
df_mandarin = questions[Mandarin]

Malay = pd.Series(df["Language"] == 'Malay')
df_malay = questions[Malay]

Tamil = pd.Series(df["Language"] == 'Tamil')
df_tamil = questions[Tamil]


#  Mean
print("Mandarin Mean: ")
print(df_mandarin.mean())
print("\nMalay Mean: ")
print(df_malay.mean())
print("\nTamil Mean: ")
print(df_tamil.mean())

#  Sample Standard Deviation
print("\nMandarin Sample Standard Deviation: ")
print(df_mandarin.std())
print("\nMalay Sample Standard Deviation: ")
print(df_malay.std())
print("\nTamil Sample Standard Deviation: ")
print(df_tamil.std())
```
